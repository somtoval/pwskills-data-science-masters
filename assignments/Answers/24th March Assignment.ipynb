### Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./../../../course/practice/Exploratory Data Analysis/resources/winequality-red.csv')
df.head()
```

- Fixed Acidity: This refers to the concentration of non-volatile acids in the wine, primarily tartaric acid. It contributes to the wine's overall acidity and affects its taste and stability. Acidity is an important factor in determining the wine's freshness and balance.

- Volatile Acidity: This feature represents the concentration of volatile acids in the wine, primarily acetic acid. Too much volatile acidity can result in a vinegary taste, which is undesirable in quality wines. Thus, a low level of volatile acidity is generally preferred.

- Citric Acid: Citric acid is a naturally occurring acid found in wine grapes. It can contribute to the wine's freshness, enhance flavors, and act as a preservative. A higher level of citric acid can be beneficial for the overall quality of the wine.

- Residual Sugar: This feature refers to the amount of sugar left in the wine after fermentation is complete. It affects the wine's perceived sweetness. Different wine styles have varying levels of residual sugar, and the optimal amount depends on the wine type and desired taste.

- Chlorides: The concentration of chlorides in wine can influence its taste and mouthfeel. Higher levels of chlorides may result in a salty or briny taste, which is generally undesirable. Moderation in chloride content is important for achieving a balanced flavor profile.

- Free Sulfur Dioxide: Sulfur dioxide is commonly used in winemaking as a preservative and antioxidant. The free sulfur dioxide content represents the amount of unbound or active sulfur dioxide in the wine. It helps prevent oxidation and microbial spoilage, thus preserving the wine's freshness and quality.

- Total Sulfur Dioxide: This feature indicates the total concentration of both free and bound sulfur dioxide in the wine. While sulfur dioxide is necessary for wine preservation, excessive levels can negatively impact the wine's aroma and taste. Therefore, maintaining an appropriate total sulfur dioxide level is crucial.

- Density: Density is a measure of the wine's mass per unit volume. It provides an indication of the wine's body or mouthfeel. Density can be influenced by factors such as sugar content, alcohol level, and dissolved solids. It contributes to the wine's perceived texture and overall sensory experience.

- pH: pH is a measure of acidity or alkalinity on a logarithmic scale. It affects the wine's stability, microbial activity, and sensory attributes. The ideal pH value varies depending on the wine style, but a lower pH generally indicates higher acidity and freshness.

- Sulphates: Sulphates, or sulfates, are a type of salt often added to wines as a preservative. They can also contribute to the wine's flavor and mouthfeel. However, excessive levels of sulfates may result in a bitter taste. The appropriate level of sulfates depends on the specific wine and winemaking practices.

- Alcohol: The alcohol content in wine plays a significant role in its flavor, body, and overall sensory perception. It affects the wine's mouthfeel, viscosity, and perceived warmth. Alcohol content is generally considered an essential factor in wine quality and balance.

- Quality (Target variable): The quality feature represents the overall quality of the wine as rated by experts or tasters. It is often a subjective assessment based on various sensory characteristics such as taste, aroma, balance, complexity, and overall enjoyment.

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import warnings
warnings.simplefilter('ignore')
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load dataset
wine = pd.read_csv('./../../../course/practice/Exploratory Data Analysis/resources/winequality-red.csv')

# Separate features and target
X = wine.drop('quality', axis=1)
y = wine['quality']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Proportion of variance explained
pca.explained_variance_ratio_
```

```python
# Cumulative proportion of variance explained
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
cumulative_variance
```

```python
# Number of components required to explain 90% of variance
n_components = np.argmax(cumulative_variance >= 0.9) + 1
print('Number of components required to explain 90% of variance:', n_components)
```
