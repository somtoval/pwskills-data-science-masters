### Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present your findings to your team?

```python
df.head()
```

```python
df['Date_of_Journey']
```

```python
df['Date'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
df['Date']
```

```python
df_monthly = df.groupby(pd.Grouper(key='Date', freq='M')).mean()
df_monthly 
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('Months')
plt.ylabel('Price')
plt.title('Price trends according to months')
plt.plot(df_monthly.index,df_monthly['Price'],color='blue',linewidth=4, linestyle='dotted', marker='D', markersize=10, alpha=0.7)
```

We can notice a price hike and fall across seasons

### Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=df.Category,y=df.Rating,data=df,palette='Set2')
plt.title('App rating by category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

As you can see business, communication, dating, finance, game, family, tools productivity has most outliers so we can say that these are the categories that affect the ratings.
